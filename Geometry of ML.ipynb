```python
import matplotlib.pyplot as plt
import numpy as np
w1, w2, b = 3, 4, 5
def line(x1, x2):
    return w1 * x1 + w2 * x2 + b

def line_x1(x1):
    return (w1* x1 + b) / (-w2)

def logit(value):
    return 1 / (1 + np.exp(-value))
 
np.random.seed(0)
x1x2 = np.random.randn(200, 2) * 2

for x1, x2 in x1x2:
    value = line(x1, x2)
    
    if (value == 0):
        plt.plot(x1, x2, 'ro', color='blue')
    elif (value > 0):
        plt.plot(x1, x2, 'ro', color='green')
    elif (value < 0):
        plt.plot(x1, x2, 'ro', color='red')
plt.plot(w1, w2, 'ro', color='black')

plt.gca().set_aspect('equal', adjustable='box')

x1_range = np.arange(-5.0, 5.0, 0.5)
plt.plot(x1_range, line_x1(x1_range), color='blue')

plt.xlabel('x1')
plt.xlabel('x2')

plt.show()
```

```python
for x1, x2 in x1x2:
    value = logit(line(x1, x2)/2)
    if (value < 0.00001):
        color = 'red'
    elif (value > 0.99999):
        color = 'green'
    else:
        color = str( 0.75 - value * 0.5)
    plt.plot( x1, x2, 'ro', color=color)

plt.gca().set_aspect('equal', adjustable='box')

plt.plot(x1_range, line_x1(x1_range), color='blue')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
x = [[-3, -5], [2, -3], [1.2, -3], [-5, -5]]
b = [-8, 4, 4, -8]
def line1(w1, w2):
    return x[0][0] * w1 + x[0][1] * w2 + b[0]

def line1_w1(w1):
    return (x[0][0] * w1 + b[0]) / (-x[0][1])

def line2(w1, w2):
    return x[1][0] * w1 + x[1][1] * w2 + b[1]
def line2_w1(w1):
    return (x[1][0] * w1 + b[1]) / (-x[1][1])

def line3(w1, w2):
    return x[2][0] * w1 + x[2][1] * w2 + b[2]
def line3_w1(w1):
    return (x[2][0] * w1 + b[3]) / (-x[2][1])

def line4(w1, w2):
    return x[3][0] * w1 + x[3][1] * w2 + b[3]

def line4_w1(w1):
    return (x[3][0] * w1 + b[3]) / (-x[3][1])

w1_range = np.arange(-5.0, 5.0, 0.5)
w2_range = np.arange(-5.0, 5.0, 0.5)

for w1 in w1_range:
    for w2 in w2_range:
        val1 = line1(w1, w2)
        val2 = line2(w1, w2)
        val3 = line3(w1, w2)
        val4 = line4(w1, w2)
        
        if (val1 <0 and val2 > 0 and val3 > 0 and val4 < 0):
            color = 'green'
        else:
            color = 'red'
        plt.plot( w1,w2, 'ro', color=color)
# рисуем саму линию (гиперплоскость) для образца 1
plt.plot(w1_range, line1_w1(w1_range), color='blue')
# для образца 2
plt.plot(w1_range, line2_w1(w1_range), color='blue')
# для образца 3
plt.plot(w1_range, line3_w1(w1_range), color='blue')
# для образца 4
plt.plot(w1_range, line4_w1(w1_range), color='blue')

# рисуем только эту область — остальное не интересно
plt.axis([-7, 7, -7, 7])

# проставляем названия осей
plt.xlabel('w1')
plt.ylabel('w2')

# на экран!
plt.show()
```
